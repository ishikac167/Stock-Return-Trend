```python
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
djia = pd.read_csv('^DJI.csv')
```

```python
djia.head()
```

```python
djia = djia.set_index('Date')
```

```python
djia.head()
```

```python
return1 = pd.DataFrame({'DJAI' : djia['Adj Close']})
return1.head()
```

```python
return1 = return1.apply(lambda x: x / x[0])
return1.head()
```

```python
plt.figure()
for i, col in enumerate(return1.columns):
    return1[col].plot()
    
plt.title("Stock Return - Dow Jones Industrial Average")
plt.axhline(y=1, color='black', linestyle='-')
plt.xlabel('Year-Month')
plt.xticks(rotation=45)
```

```python
import pandas_datareader as web
import datetime
start = datetime.datetime(2019,12,6)
end = datetime.datetime(2020,12,4)
apple = web.DataReader("AAPL", "yahoo", start, end)
microsoft = web.DataReader("MSFT", "yahoo", start, end)
google = web.DataReader("GOOG", "yahoo", start, end)
tesla = web.DataReader("TSLA", "yahoo", start, end)
ibm = web.DataReader("IBM", "yahoo", start, end)
```

```python
stocks = pd.DataFrame({"AAPL": apple["Adj Close"],
            "MSFT": microsoft["Adj Close"],
                      "GOOG": google["Adj Close"],
                      "TSLA" : tesla["Adj Close"],
                      "IBM" : ibm["Adj Close"]})
stocks.head()
```

```python
stocks = pd.DataFrame({"AAPL": apple["Adj Close"],
            "MSFT": microsoft["Adj Close"],
                      "GOOG": google["Adj Close"],
                      "TSLA" : tesla["Adj Close"],
                      "IBM" : ibm["Adj Close"],
                      "DJIA" : djia["Adj Close"]})
stocks.head()
```

```python
stock_return = stocks.apply(lambda x : x/x[0])
stock_return.head()
```

```python
plt.figure()
for i,col in enumerate(stock_return.columns):
    stock_return[col].plot()
    
plt.title('Stock Return Trend over a year with DJIA')
plt.xlabel('Year-Month')
plt.xticks(rotation=45)
plt.legend(stock_return.columns)
plt.axhline(y=1, color='black', linestyle='-')
```

```python
stock_return.tail()
```
